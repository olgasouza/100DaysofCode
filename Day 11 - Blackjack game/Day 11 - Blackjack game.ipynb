```python
# DAY 11 - CAPSTONE PROJECT ONE - BLACKJACK GAME 
# Made with the tips provided by Angela Yu

import random
from subprocess import call
import os
from art import logo

# to clear the screen:
def clear():
    _ = call('clear' if os.name == 'posix' else 'cls')

def deal_card():
    cards = [11, 2, 3, 4, 5, 6, 7, 8, 9, 10, 10, 10, 10]
    card = random.choice(cards)
    return card

def calculate_score(cards_list):
    # beginning of game - is any player lucky enough to get a blackjack in the first round?
    if sum(cards_list) == 21 and len(cards_list) == 2:
        return 21
    
    if sum(cards_list) > 21 and 11 in cards_list:
        cards_list.remove(11)
        cards_list.append(1)

    return sum(cards_list)

def compare(user_score, computer_score):
    if user_score > 21 and computer_score > 21:
        return "You both went over 21 :("

    if user_score == 21:
        return "You win with a blackjack! :D"
    elif computer_score == 21:
        return "You lost, computer has a blackjack :("
    elif user_score == computer_score:
        return "It's a draw!"
    elif user_score > 21:
        return "You lost, you went over 21 :("
    elif computer_score > 21:
        return "You win, computer went over 21! :D"
    elif user_score > computer_score:
        return "You win, you have the highest sum! :D"
    else:
        return "You lost :("


def play_game():
    print(logo)
    user_cards = []
    computer_cards = []
    # append the first two cards
    for _ in range(2):
        user_cards.append(deal_card())
        computer_cards.append(deal_card())

    is_game_over = False
    while not is_game_over:
        user_score = calculate_score(user_cards)
        computer_score = calculate_score(computer_cards)
        print(f"Your cards: {user_cards}, total sum: {user_score}")
        print(f"Computer's first card: {computer_cards[0]}")
        
        if user_score == 21 or computer_score == 21 or user_score > 21:
            is_game_over = True

        else:
            draw_again = input("\nDo you want to draw a new card? y/n: ").lower()
            if draw_again == 'y':
                user_cards.append(deal_card())
            else:
                is_game_over = True

    while computer_score != 0 and computer_score < 16:
        computer_cards.append(deal_card())
        computer_score = calculate_score(computer_cards)
        
    print(f"\nYour cards: {user_cards}, total sum: {user_score}")
    print(f"Computer's cards: {computer_cards}, total sum: {computer_score}")
    print(compare(user_score, computer_score))

# to keep playing:
while input("\nDo you want to play a game of blackjack? y/n: ").lower() == 'y':
    clear()
    play_game()

print("Goodbye then")
```
